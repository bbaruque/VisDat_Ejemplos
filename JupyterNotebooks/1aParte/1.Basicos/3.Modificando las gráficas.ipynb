# Modificando las gráficas

Al presentar las gráficas a una audiencia externa, es importante modificar los estilos de la gráfica. Normalmente las opciones por defecto no son las más adecuadas. Esta es una tarea importante para aumentar el poder comunicativo de las visualizaciones y remarcar aquellos aspectos que puedan resultar más interesantes.

Veremos como se puede modificar aspectos del estilo del gráfico que estamos creando. Prácticamente todos los elementos visuales del gráfico son susceptibles de ser modificados, por lo que no se intentará mostrar todos. Se intentará mostrar un ejemplo de las operaciones más comunes y útiles: ajustar el tamaño de la figura, los colores y tamaños de fuentes, añadir títulos y eliminar o incluir bordes y ejes.

Para llegar a dominar bien las herramientas, es imprescindible revisar la documentación y foros profesionales del estilo de [StackOverflow](https://stackoverflow.com/search?q=pandas+plot). 

- Ya se ha mencionado anteriormente la documentación de `pandas`: https://pandas.pydata.org/pandas-docs/stable/visualization.html
-  En este tutorial se utiliza tambien la librería `matplotlib`: https://matplotlib.org/index.html (Es en la que se apoya la librería de Pandas)

```python
import pandas as pd
```

```python
# En caso de haber ejecutado la celda anterior, esta debe dejarse SIN EJECUTAR
data_dir = "../../data/datos_1ªy2ªParte/"
```

```python
reviews = pd.read_csv(data_dir+"/winemag-data_first150k.csv", index_col=0)
reviews.head(3)
```

## Opciones por defecto

```python
reviews['points'].value_counts().sort_index().plot.bar()
```

## Tamaño del gráfico

El parametro `figsize` controla el tamaño de la imagen final (en pulgadas). Necesita recibir como parámetro una tupla indicando el ancho y el alto.

```python
reviews['points'].value_counts().sort_index().plot.bar(figsize=(12, 6))
```

## Color de las barras

El parametro `color` controla el color de las barras.

```python
reviews['points'].value_counts().sort_index().plot.bar(
    figsize=(12, 6),
    color='mediumvioletred'
)
```

## Tamaño del texto (etiquetas)

El parametro `fontsize` controla el tamaño de las etiquetas.

```python
reviews['points'].value_counts().sort_index().plot.bar(
    figsize=(12, 6),
    color='mediumvioletred',
    fontsize=16
)
```

## Título

El parametro `title` permite especifiar un título en la cabecera del gráfico.

```python
reviews['points'].value_counts().sort_index().plot.bar(
    figsize=(12, 6),
    color='mediumvioletred',
    fontsize=16,
    title='Rankings Given by Wine Magazine',
)
```

## Librerías adicionales

### Tamaño del texto (título)

En este caso, deberemos ajustarlo de manera menos directa.

Es importante saber que Pandas emplea para realizar los gráficos la librería de bajo nivel [`matplotlib`](https://matplotlib.org/).
Podemos utilizar las utilidades de ésta para realizar modificaciones más finas. Cualquier gráfico u objeto que se pueda crear  con pandas, se podría hacer también con `matplotlib` (quizá en más pasos).

Empleando `matplotlib`, podemos cambiar características de gráficos cosntruidos con Pandas:

```python
import matplotlib.pyplot as plt

# Se genera el gráfico inicial y se almacena una referencia al mismo en 'ax'
ax = reviews['points'].value_counts().sort_index().plot.bar(
    figsize=(12, 6),
    color='mediumvioletred',
    fontsize=16
)
# Se emplea la referencia para modificar el título
ax.set_title("Rankings Given by Wine Magazine", fontsize=20)
```

En el código anterior, lo que se hace es crear una variable que permite almacenar una referencia a los ejes del gráfico que se ha generado. 

Estos ejes son un objeto generado por la librería `matplotlib` (a la que llama `pandas` de forma transparente a nosotros). Se emplean a continuación para incluir un título sobre el gráfico (modificando el tamaño). Para ello podemos ahora emplear un comando de `matplotlib`, que permite más modificaciones que las típicas de `pandas`.

### Modificación de Ejes

Seaborn es otra librería que emplea `matplotlib` como base y permite realizar otro tipo de acciones sobre las gráficas (veremos más de ella en adelante).

```python
import matplotlib.pyplot as plt
import seaborn as sns

ax = reviews['points'].value_counts().sort_index().plot.bar(
    figsize=(12, 6),
    color='mediumvioletred',
    fontsize=16
)
ax.set_title("Rankings Given by Wine Magazine", fontsize=20)
sns.despine(bottom=True, left=True)
```
